This project focuses on exploring and visualizing the famous Iris dataset. 
The objective is to understand the structure of the dataset, analyze feature 
relationships, and visualize patterns using Python data science libraries.

### **Tools Used**
- Python
- Pandas
- Matplotlib
- Seaborn

### **1- Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **2- Load the Dataset**

```python
iris = sns.load_dataset("iris")
```

### **5- Scatter Plot (Relationship Between Variables)**

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Relationship Between Sepal Length and Petal Length")
plt.show()
```

### **6- Histogram (Distribution)**

```python
plt.figure(figsize=(8,6))

sns.histplot(iris["sepal_length"], bins=20)

plt.title("Distribution of Sepal Length")
plt.show()
```

### **7- Box Plot (Outlier Detection)**

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=iris,
    x="species",
    y="petal_length"
)

plt.title("Petal Length Distribution Across Species")
plt.show()
```
